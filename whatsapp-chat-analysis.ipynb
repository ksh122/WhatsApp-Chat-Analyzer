```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with Friendzone.txt','r', encoding='utf-8')
```

```python
data = f.read() # stores data in form of string
```

```python
print(data)
```

```python
# pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
pattern = '\d{2}\/\d{2}\/\d{4},\s\d{2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
# convert message_data type
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %H:%M - ')

df.rename(columns={'message_date': 'date'}, inplace=True)

df.head()
```

```python
df.shape
```

```python
# separate users and messages
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message, 1)
    if len(entry)>2: # If valid user-message pattern exists
        users.append(entry[1].strip())
        messages.append(entry[2].strip())
    else:
        users.append('group_notification')
        messages.append(entry[0].strip()) # Entire message as it's a system message

df['user'] = users
df['message'] = messages
# df.drop(columns=['user_message'], inplace = True)

df.sample(8)
```

```python
df['year']= df['date'].dt.year
df['month']= df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute']= df['date'].dt.minute
```

```python
# remove group messages
# remove media omitted message
# remove stopwords
```

```python
temp = df[df['user'] != 'group_notification']
temp = temp[temp['message'] != '<Media omitted>']
```

```python
f = open('stop_hinglish.txt','r')
stop_words = f.read()
print(stop_words)
```

```python
words = []

for message in temp['message']:
    for word in message.lower().split():
        if word not in stop_words:
            words.append(word)

from collections import Counter
pd.DataFrame(Counter(words).most_common(20))
```
